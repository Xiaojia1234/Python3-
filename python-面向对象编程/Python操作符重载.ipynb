# Python操作符重载

可以根据所使用的操作数更改python张运算符的含义。这种做法被称为运算符重载。

python操作系统适用于内置类。但统一运算符的行为在不同的类型有所不同。例如，+运算符将对两个数字执行算术加法，
合并两个列表并连接两个字符串。

python中的这个功能允许相同的操作复根据上下文的不同，其含义称为运算符重载。

那么当将它们与用户定义的类的对象一起使用时会发生什么？考虑下面的类，它试图模拟二维坐标系中的一个点。

```python
class Point:
    def __init__(self, x = 0, y = 0):
        self.x = x
        self.y = y
```

现在，运行代码，尝试在Python shell中添加两点。

```python
p1 = Point(2,3)
p2 = Point(-1,2)
p1+p2  # 会报错
```

### Python中的特殊功能

```python
p1 = Point(2,3)
print(p1)
```

```python
class  Point:
    def __init__(self,x = 0,y = 0):
        self.x = x
        self.y = y
    
    def __str__(self):
        return("{0},{1}".format(self.x,self.y))
```

```python
p1 = Point(4,5)
print(p1)
```

当使用内置函数str()或format()时，调用同样的方法。

```python
str(p1)
```

```python
format(p1)
```

### 在Python中重载+运算符

```python
class Point:
    def __init__(self,x=0,y=0):
        self.x=x
        self.y=y
    
    def __str__(self):
        return("{0},{1}".format(self.x,self.y))
    
    def __add__(self,other):
        x = self.x +other.x
        y = self.y + other.y
        return Point(x,y)
```

```python
p1 = Point(2,3)
p2 = Point(-1,2)
print(p1+p2)
```

```python
print(p1.__add__(p2))
```

### 在Python中重载比较运算符

python不会限制操作符重载运算符。也可以承载比较运算符。

假设想在Point类中实现小于符号<比较运算。

比较这些来自原点的数值，并为此返回结果。可以实现如下：

```python
class Point:
    def __init__(self, x = 0, y = 0):
        self.x = x
        self.y = y

    def __str__(self):
        return "({0},{1})".format(self.x,self.y)

    def __lt__(self,other):
        self_mag = (self.x ** 2) + (self.y ** 2)
        other_mag = (other.x ** 2) + (other.y ** 2)
        return self_mag < other_mag
```

在Python shell中尝试这些示例运行

```python
Point(1,1)<Point(-2,-3)
```

```python
Point(1,1) < Point(0.5,-0.2)
```

```python
Point(1,1) < Point(1,1)
```
